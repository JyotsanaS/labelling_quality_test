```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine

df = df_updated.copy()

# Assuming df is your dataframe with columns: filepath, embedding, product_id

# Step 1: Convert the 'embedding' column from lists to numpy arrays
df['embedding'] = df['embedding'].apply(np.array)
df['identified_product_no'] = df['identified_product_no'].fillna(-1).astype(int).astype(str)

# Calculate the centroid for each class
class_centroids = df.groupby('identified_product_no')['embedding'].apply(lambda x: np.mean(np.vstack(x), axis=0))

# Step 2: Calculate the distance of each product's embedding from its class centroid
df['distance_to_centroid'] = df.apply(lambda row: cosine(row['embedding'], class_centroids[row['identified_product_no']]), axis=1)
```
